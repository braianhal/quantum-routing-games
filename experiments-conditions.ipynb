## Experimentos: condiciones para que los protocolos cuánticos superen a CM.

Imports.

```python
import src.game_builder as gb

import utils.chart as chart
import utils.metric as mu
from utils.network import NetworkDrawer 
import utils.tests as test

import matplotlib.pyplot as plt
import numpy as np
import random
```

Configuración de semilla.

```python
RANDOM_SEED = 702
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
```

### 1. Incremento en complejidad de la red para igual n y m.

Configuración de experimentos.

```python
rounds = 100
protocols = [
    (gb.cm(), "CM"),
    (gb.mwp(), "MWP"),
    (gb.mwm(), "MWM")]
test_cases_3 = test.combinations_test_cases_3(8)
test_cases_4 = test.combinations_test_cases_4(8)
test_cases_5 = test.combinations_test_cases_5(8)
tests_per_case = 5
```

Generación de gráficos.

```python
def plot_results(execution_metrics):
    for metric, metric_label in mu.MEAN_METRICS:
        chart.pie(execution_metrics, metric, [p[1] for p in protocols], metric_label)
```

Ejecución de experimentos.

```python
execution_metrics = []
for test_cases in [test_cases_3, test_cases_4, test_cases_5]:
    results = test.execute_combinations_test(rounds, test_cases, tests_per_case, [p[0] for p in protocols])
    execution_metrics.append(results)
```

Compilar resultados.

```python
results_v3 = execution_metrics[0] 
results_v4 = execution_metrics[1]
results_v5 = execution_metrics[2]
```

Resultados para $v = 3$:

```python
print("Cantidad casos:", len(results_v3))
plot_results(results_v3)
```

Resultados para $v = 4$:

```python
print("Cantidad casos:", len(results_v4))
plot_results(results_v4)
```

Resultados para $v = 5$:

```python
print("Cantidad casos:", len(results_v5))
plot_results(results_v5)
```
