```python
from network_utils import NetworkDrawer, NetworkGenerator
from classical_routing_protocols import RandomPathProtocol, BestLatencyProtocol, OptimizableLatencyProtocol 
from routing_protocols import OptimalFlowProtocol
import random

random.seed = 1234
```

```python
networkGenerator = NetworkGenerator()
drawer = NetworkDrawer()
```

```python
N = networkGenerator.generate_random(5)
```

```python
drawer.draw(N, "latency")
```

```python
N, metrics = OptimalFlowProtocol().run(N, 100)
drawer.draw(N, "flow")
print(metrics)
```

```python
N, metrics = RandomPathProtocol().run(N, 100, 100)
drawer.draw(N, "flow")
print(metrics)
```

```python
N, metrics = BestLatencyProtocol().run(N, 100, 100)
drawer.draw(N, "flow")
print(metrics)
```

```python
N, metrics = OptimizableLatencyProtocol(5).run(N, 100, 100)
drawer.draw(N, "flow")
print(metrics)
```
