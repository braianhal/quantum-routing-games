## Experimentos: hiperparámetro $\sigma$.

Imports.

```python
import src.game_builder as gb

import utils.chart as chart
import utils.metric as mu
import utils.tests as test

import matplotlib.pyplot as plt
import numpy as np
import random
```

Configuración de semilla.

```python
RANDOM_SEED = 937
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
```

Configuración de experimentos.

```python
test_cases = test.get_test_cases(10)
rounds = 100
protocol_providers = [
    lambda sigma : gb.ewl(sigma = sigma, n_params_per_qubit = 1),
    lambda sigma : gb.ewl(sigma = sigma, n_params_per_qubit = 2),
    lambda sigma : gb.ewl(sigma = sigma, n_params_per_qubit = 3)]
sigmas = np.arange(0.05,2.01,0.05)
```

Ejecución de experimentos.

```python
execution_metrics = test.execute_hyperparameter_test(rounds, test_cases, protocol_providers, sigmas)
```

Visualización de resultados.

```python
for metric, metric_label in mu.MEAN_METRICS:
    chart.series(sigmas, [[m[metric] for m in pm] for pm in execution_metrics], [p(1).get_name() + "("+str(i+1)+")" for p,i in zip(protocol_providers, range(len(protocol_providers)))], (r"$\sigma$", metric_label), 
                 window=1, optimal=1 if ("poa" in metric) else None, show_mins=True, figsize=(8,4))
```
